#Experiment 2: Creating and Printing Tensors

```python
import tensorflow as tf
print("TensorFlow Version:", tf.__version__)
```

```python
scalar = tf.constant(10)
print("Scalar : ",scalar)

vector = tf.constant([1, 2, 3, 4])
print("Vector : " ,vector)

matrix = tf.constant([[1, 2], [3, 4]])
print("Matrix : ",matrix)

t3=tf.constant([[[1,2],[3,4]],[[5,6],[7,8]]])
print("Three Dimensional Scalar :", t3)
```

#Experiment 3: Basic Tensor Operations

```python
a = tf.constant(5)
b = tf.constant(2)

print("=== Scalar Operations ===")
print("Addition:", tf.add(a, b))
print("Subtraction:", tf.subtract(a, b))
print("Multiplication:", tf.multiply(a, b))
print("Division:", tf.divide(a, b))
print("Power (a^b):", tf.pow(a, b))

```

```python
a = tf.constant([5, 10, 15])
b = tf.constant([1, 2, 3])

print("=== Vector Operations ===")
print("Addition:", tf.add(a, b))
print("Subtraction:", tf.subtract(a, b))
print("Multiplication:", tf.multiply(a, b))
print("Division:", tf.divide(a, b))
print("Power (a^b):", tf.pow(a, b))
```

```python
a = tf.constant([[2, 4],
                 [6, 8]])

b = tf.constant([[1, 2],
                 [3, 4]])

print("=== Matrix Operations ===")
print("Addition:\n", tf.add(a, b))
print("Subtraction:\n", tf.subtract(a, b))
print("Multiplication (element-wise):\n", tf.multiply(a, b))
print("Division:\n", tf.divide(a, b))
print("Power (a^b):\n", tf.pow(a, b))
```

#Experiment 4: Data Types and Shape of Tensors

```python
a = tf.constant([[1, 2, 3],
                 [4, 5, 6]])

print("Tensor:", a)
print("Data Type:", a.dtype)
print("Shape:", a.shape)
print("Number of Dimensions:", a.ndim)
```

#Experiment 5: Converting NumPy Arrays to Tensors

```python
import numpy as np
import tensorflow as tf

# NumPy array
a = np.array([10, 20, 30, 40])

# Convert to Tensor
b = tf.convert_to_tensor(a)

print("NumPy Array:", a)
print("TensorFlow Tensor:", b)
```
